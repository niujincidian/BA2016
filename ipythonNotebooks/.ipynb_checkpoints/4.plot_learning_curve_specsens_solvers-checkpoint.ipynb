```python
#####################plot#####################
K = len(soundTypes)  #get number of class labels

#get y aixs range
accValues = np.concatenate((sens_SGD,spec_SGD,bal_SGD,sens_Adam,spec_Adam,bal_Adam))
minY = np.min(accValues)
maxY = np.max(accValues)
for s in range(K):
    plt.figure(figsize=(20,6))
    plt.suptitle('SigmoidCrossEntropy '+soundTypes[s],fontsize=24)
    
#     plt.plot(it_list_SCE, sens_list_SCE[:,s],color='green',label='sens')
    plt.plot(it_SGD, sens_SGD[:,s],color='green',label='sens-SGD')
    plt.plot(it_Adam, sens_Adam[:,s],'--',color='green',label='sens-Adam')

    
#     plt.plot(it_list_SCE, spec_list_SCE[:,s], color='red',label='spec')
    plt.plot(it_SGD, spec_SGD[:,s], color='red',label='spec-SGD')
    plt.plot(it_Adam, spec_Adam[:,s],'--',color='red',label='spec-Adam')
    
#     plt.plot(it_list_SCE, bal_acclist_SCE[:,s],color='blue',label='bal')
    plt.plot(it_SGD, bal_SGD[:,s],color='blue',label='bal-SGD')
    plt.plot(it_Adam, bal_Adam[:,s],'--',color='blue',label='bal-Adam')
    
    plt.xlabel('iteration')
    plt.ylabel('sens-spec')
    
    plt.ylim((minY,maxY))
    ticks, _ = plt.xticks()
    
    plt.xticks(ticks, ["%dK" % int(t/1000) for t in ticks])

#     plt.legend(loc=0,prop={'size':15})
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,prop={'size':18})

    ########## TODO ###########
    # save the plots under the directory you choose
#     plt.savefig(root_path+'acc_measures'+str(s+1).zfill(2)+soundTypes[s]+'.png')

    # print the values from last iteration
    print 'last iteration acc-measurements for %s'%soundTypes[s]
#     print '\t SCE \t SCE_ndrp '
    print '\t SGD \t \t Adam '
    print 'sens\t %f \t %f'%(sens_SGD[-1,s],sens_Adam[-1,s])
    print 'spec \t %f \t %f'%(spec_SGD[-1,s],spec_Adam[-1,s])
    print 'bal \t %f \t %f'%(bal_SGD[-1,s],bal_Adam[-1,s])
```

### average accuracy measure

```python
avSensSGD = np.mean(sens_SGD,axis=1)
avSpecSGD = np.mean(spec_SGD,axis=1)
avBalSGD = np.mean(bal_SGD,axis=1)

avSensAdam = np.mean(sens_Adam,axis=1)
avSpecAdam = np.mean(spec_Adam,axis=1)
avBalAdam = np.mean(bal_Adam,axis=1)

accValues = np.concatenate((avSensSGD,
                           avSpecSGD,
                           avBalSGD,
                           avSensAdam,
                          avSpecAdam,
                          avBalAdam))
minY = np.min(accValues)
maxY = np.max(accValues)
plt.figure(figsize=(20,6))
plt.suptitle('Average Performance',fontsize=24)
    
plt.plot(it_SGD, avSensSGD,color='green',label='sens-SGD')
plt.plot(it_Adam, avSensAdam,'--',color='green',label='sens-Adam')

    
plt.plot(it_SGD, avSpecSGD, color='red',label='spec-SGD')
plt.plot(it_Adam, avSpecAdam,'--',color='red',label='spec-Adam')
    
plt.plot(it_SGD, avBalSGD,color='blue',label='bal-SGD')
plt.plot(it_Adam, avBalAdam,'--',color='blue',label='bal-Adam')
    
plt.xlabel('iteration')
plt.ylabel('sens-spec')
    
plt.ylim((minY,maxY))

ticks, _ = plt.xticks()
    
plt.xticks(ticks, ["%dK" % int(t/1000) for t in ticks])

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,prop={'size':18})
plt.show()

######print average#######
print 'average balanced accuracy'
print '\t SGD \t \t Adam '
print 'sens\t %f \t %f'%(avSensSGD[-1],avSensAdam[-1])
print 'spec \t %f \t %f'%(avSpecSGD[-1],avSpecAdam[-1])
print 'bal \t %f \t %f'%(avBalSGD[-1],avBalAdam[-1])
```
